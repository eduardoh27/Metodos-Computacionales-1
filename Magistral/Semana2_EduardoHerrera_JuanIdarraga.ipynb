```python
!pip install numpy 
!pip install matplotlib 
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as anim

trayectorias = []
# Random walk

N = 1000 # 1000 pasos

for particula in range(100): # 100 part√≠culas
    
    x = np.zeros(N)
    y = np.zeros(N)

    for i in range(1,N):

        theta = 2*np.pi*np.random.rand() # Genera un numero aleatorio entre 0 y 2pi

        # Futuro # Presente

        x[i] = x[i-1] + np.cos(theta)
        y[i] = y[i-1] + np.sin(theta)
        
    tray = {"x": x, "y": y}
    trayectorias.append(tray)


    
# Definamos la animacion

fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(1,1,1)
# numero de filas, numero de columnas, posicion en el array

def init():
    ax.set_xlim(-15,15)
    ax.set_ylim(-15,15)
    ax.set_xlabel(r'$x[m]$',fontsize=15)
    ax.set_ylabel(r'$y[m]$',fontsize=15)
    
# Definimos una funcion de actualizacion

def Update(i):
    plot = ax.clear()
    init()
    for trayectoria in trayectorias:
        x = trayectoria["x"]
        y = trayectoria["y"]
        plot = ax.plot(x[i],y[i],color='k',marker='o')
    return plot

Animation = anim.FuncAnimation(fig,Update,frames=N,init_func=init)
```
